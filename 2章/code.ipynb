```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import factorial
%matplotlib inline
plt.style.use('seaborn-white')
plt.rcParams['figure.figsize'] = (5,3)
```

```python
data = [2, 2, 4, 6, 4, 5, 2, 3, 1, 2, 0, 4, 3, 3, 3, 3, 4, 2, 7, 2, 4, 3, 3, 3, 4, 3, 7, 5,
        3, 1, 7, 6, 4, 6, 5, 2, 4, 7, 2, 2, 6, 2, 4, 5, 4, 5, 1, 3, 2, 3]
```

```python
len(data)
```

```python
#summary statistics
data = np.array(data)
data = pd.Series(data)
data.describe()
```

```python
#ヒストグラム
plt.hist(data, color='white', edgecolor='black', bins=8)
plt.show()
```

```python
#標本分散
print(data.var())
#標本標準偏差
print(data.std())
print(np.sqrt(data.var()))
```

```python
# 平均3.56のポアソン分布にyを当てはめる
y = np.arange(0, 10)
poisson = np.exp(-3.56) * np.power(3.56, y) / factorial(y)
# プロット
plt.plot(y, poisson, 'o-', color='black')
plt.hist(data, bins=8, range=(0, 8), color='white', edgecolor='black', density=True)
plt.show()
```

```python
#対数尤度logL(λ)とλの関係
def logL(lam):
    return np.sum(np.log(np.exp(-lam) * np.power(lam, data) / factorial(data)))
lam = np.linspace(2, 5, 100)
logL = [logL(l) for l in lam]
plt.plot(lam, logL)
plt.grid(color='k', linewidth=0.5)
plt.show()
```
